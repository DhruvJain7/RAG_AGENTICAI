```python
from langchain_community.chat_message_histories import ChatMessageHistory
```

```python
from langchain_ollama import ChatOllama

chat_llm = ChatOllama(
    model = "llama3",
    temperature = 0.8 ,
    num_predict = 256,
)
```

```python
history = ChatMessageHistory()
```

```python
history.add_ai_message("hi!")
```

```python
history.add_user_message("what is the capital of France?")
```

```python
history.messages
```

```python
ai_response = chat_llm.invoke(history.messages)
print(ai_response.content)
```

#### Conversation buffer
Conversation buffer memory allows for the storage of messages, which you use to extract messages to a variable. Consider using conversation buffer memory in a chain, setting `verbose=True` so that the prompt is visible. 

```python
from langchain_classic.chains.conversation.memory import ConversationBufferMemory
from langchain_classic.chains import ConversationChain 

conversation = ConversationChain(
    llm = chat_llm,
    verbose = True,
    memory = ConversationBufferMemory(
)
```
